```python
with open("./Chr4.seq", "r") as fp:
    seq = fp.read()
print(len(seq))
```

```python
CG = 0
for i in range(1, len(seq)):
    if seq[i] == "C" or seq[i] == "G":
        CG += 1
print((100*CG)/float(len(seq)-1))
    
```

```python
def get_allele_percentage(allele, seq):
    count = 0
    for i in range(1, len(seq)):
        if seq[i] == allele:
            count += 1
    return ((100*count)/float(len(seq)-1))
```

```python
get_allele_percentage("A", seq)
```

```python
get_allele_percentage("T", seq)
```
