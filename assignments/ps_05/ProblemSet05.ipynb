**Problem #1:** Consider the table of integer molecular weights for amino acid residues that we have used in class shown below.

        Daltons = { 
            'A':  71, 'C': 103, 'D': 115, 'E': 129,
            'F': 147, 'G':  57, 'H': 137, 'I': 113,
            'K': 128, 'L': 113, 'M': 131, 'N': 114,
            'P':  97, 'Q': 128, 'R': 156, 'S':  87,
            'T': 101, 'V':  99, 'W': 186, 'Y': 163 
        }

Which peptide *combinations* have a total molecular weight of 336? A *combination* means that the order of the peptides is unimportant (i.e. "AHK" is the same combination as any of "AKH", "KAH", "KHA", "HAK", and "HKA"). When reporting your answer you should lexigraphically sort the resides within each combination ("AHK" is the sorted version of the previous example). Furthermore, the list of combinations should also be lexigraphically sorted (The python sorted() method will do this for you).

```python
Daltons = { 
        'A':  71, 'C': 103, 'D': 115, 'E': 129,
        'F': 147, 'G':  57, 'H': 137, 'I': 113,
        'K': 128, 'L': 113, 'M': 131, 'N': 114,
        'P':  97, 'Q': 128, 'R': 156, 'S':  87,
        'T': 101, 'V':  99, 'W': 186, 'Y': 163 
    }
```

```python
combs = [] # grows at ((N*(i+N))/2), where N is the length of the combination
product = [] # grows at 3 ** N, where N is the length of the product

for i in range(10):
    combs.append(len([''.join(p) for p in itertools.combinations_with_replacement('AKH', i)]))
    product.append(len([''.join(p) for p in itertools.product('AKH', repeat=i)]))
    
print(combs)
print(product)
```

```python
import itertools

def find_weight_iterative(target): 
    combinations = set()
    for i in range(5):
        for comb in [''.join(p) for p in itertools.combinations_with_replacement("".join(acid for acid in Daltons), i)]:
            if sum(Daltons[char] for char in comb) == target:
                combinations.add(comb)
    return sorted(combinations)
```

```python
%time find_weight_iterative(336)
```

```python
def find_weight(target, prefix=""):
    global answers
    if len(prefix) == 0:
        answers = []
    mw = sum([Daltons[r] for r in prefix])
    if mw == target:
        answers.append(prefix)
        return
    elif (mw > target):
        return
    for r in sorted(Daltons):
        if len(prefix) == 0 or r >= prefix[-1]:
            find_weight(target, prefix + r)
```

```python
%time find_weight(336)
print(answers)
```

**Problem #3:** How many distinct molecular weights are in the Theoretical Spectrum of *ORF10*?  

```python
ORF10 = "MGYINVFAFPFTIYSLLLCRMNSRNYIAQVDVVNFNLT"
```

```python
def TheoreticalSpectrum(peptide):
    # Generate every possible fragment of a peptide
    spectrum = set()
    for fragLength in range(1,len(peptide)+1):
        for start in range(0,len(peptide)-fragLength+1):
            seq = peptide[start:start+fragLength]
            spectrum.add(sum([Daltons[res] for res in seq]))
    return sorted(spectrum)
```

```python
print(len(TheoreticalSpectrum(ORF10)))
```

**Problem #4:** What is the most frequently occcuring molecular weight of ORF10 subpeptides? What is the most frequently occcuring molecular weight of ORF10 of peptides composed of more than two residues? 

```python
from collections import defaultdict

def Subpeptide_weights(peptide):
    weights = defaultdict(list)
    for fragLength in range(1, len(peptide)+1):
        for start in range(0, len(peptide)-fragLength+1):
            seq = peptide[start:start+fragLength]
            mw = sum([Daltons[res] for res in seq])
            weights[mw].append(seq)
    return weights
```

```python
superDict = Subpeptide_weights(ORF10)
superDict
```

```python
max_count = 0
max_mw = 0
for mw in superDict:
    if len(superDict[mw]) > max_count:
        max_count = len(superDict[mw])
        max_mw = mw

print(max_mw, max_count, superDict[max_mw])
```

```python
max_count = 0
max_mw = 0
for mw in superDict:
    more_than_two = [p for p in superDict[mw] if len(p) > 2]
    if len(more_than_two) > max_count:
        max_count = len(more_than_two)
        max_mw = mw

print(max_mw, max_count, superDict[max_mw])
```
